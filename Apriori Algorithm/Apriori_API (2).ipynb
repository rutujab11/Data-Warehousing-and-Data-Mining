## APRIORI ALGORITHM USING API

CALCULATING SUPPORTS OF EACH CANDIDATE SETS

```python
from mlxtend.frequent_patterns import apriori
```

```python
dataset = [[1,2,5],
           [2,4],
           [2,3],
           [1,2,4],
           [1,3],
           [2,3],
           [1,3],
           [1,2,3,5],
           [1,2,3]]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
        
te = TransactionEncoder()                  #object used convert list of lists into numpy array
te_ary = te.fit(dataset).transform(dataset)#fit() identifies the label,transform() converts list of lists into boolean numpy array
te_ary
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)#conversion of boolean numpy array to tabular format
df
```

```python
from mlxtend.frequent_patterns import apriori

apriori(df, min_support=0.22, use_colnames=True)# #returns items with atleast 22% support and to print the column names
```

SELECTING AND PRUNING

```python
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)#finding the length of candidate sets to filter the results
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
#found the final candidate sets whose support>=0.22
frequent_itemsets[ (frequent_itemsets['length'] == max(frequent_itemsets.length))]            
```
