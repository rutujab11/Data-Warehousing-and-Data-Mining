```python
data=data.drop_duplicates()
data=data.dropna()
```

```python
data["Stadium"]=data["Stadium"].str.upper()
data["Stage"]=data["Stage"].str.lower()
data["Referee"]=data["Referee"].str.title()
```

```python
import matplotlib.pyplot as plt
plt.hist(data["RoundID"])
plt.xlabel("Rounds")
plt.ylabel("ID's")
plt.title("All Stadium")
plt.show()
```

```python
import seaborn as sb
x=data["Home Team Goals"]
y=data["Away Team Goals"]
ax = sb.stripplot(x, y); 
ax.set(xlabel ='Home', ylabel ='Away')
plt.title("All Goals")
plt.show()
```

```python
#Normalization
#Min_Max_Scaler
from sklearn import preprocessing
import numpy as np
a=data[["Attendance"]].values
mms=preprocessing.MinMaxScaler()
normal=mms.fit_transform(a)
data["Attendance"] = pd.DataFrame(normal)
```
